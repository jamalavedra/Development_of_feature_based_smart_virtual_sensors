```python
import sys
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
import warnings 
from ast import literal_eval
import math

warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
df_LR = pd.read_csv('../CSV_files/BALL/dataBall_LR_RESULTS.csv')
df_LSTM = pd.read_csv('../CSV_files/BALL/dataBall_LSTM_RESULTS.csv')
```

```python
volume_LR = [literal_eval(i) for i in df_LR['volume cm3']]
displacement_LR = [literal_eval(i) for i in df_LR['Displacement']]
coordinates_LR = [literal_eval(i) for i in df_LR['Coordinates']]
filtered_LR = [literal_eval(i) for i in df_LR['3_Filtered']]

volume_LSTM = [literal_eval(i) for i in df_LSTM['volume cm3']]
displacement_LSTM = [literal_eval(i) for i in df_LSTM['Displacement']]
coordinates_LSTM = [literal_eval(i) for i in df_LSTM['Coordinates']]
filtered_LSTM = [literal_eval(i) for i in df_LSTM['3_Filtered']]
```

```python
plt.plot(volume_LR[0], label = "real volume")
plt.plot(coordinates_LR[0], label = "SIFT")
plt.plot(coordinates_LR[1], label = "Convolutional features")

plt.suptitle('Volume Prediction with linear regression')
plt.xlabel('epochs')
plt.ylabel('volume (cm^3)')
plt.legend()
plt.show()
```
